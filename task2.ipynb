## Question 1: Distance Matrix Calculation

```python
def calculate_distance_matrix(df):
     """
    Calculate a distance matrix based on the dataframe, df.

    Args:
        df (pandas.DataFrame)

    Returns:
        pandas.DataFrame: Distance matrix
    """
    a # Write your logic here
    locations = set(df['id_start'].unique()).union(set(df['id_end'].unique()))
    distances = {location: {loc: float('inf') for loc in locations} for location in locations}

    for location in locations:
        distances[location][location] = 0

    for _, row in df.iterrows():
        distances[row['id_start']][row['id_end']] = row['distance']
        distances[row['id_end']][row['id_start']] = row['distance']

    for intermediate in locations:
        for source in locations:
            for target in locations:
                new_distance = distances[source][intermediate] + distances[intermediate][target]
                if new_distance < distances[source][target]:
                    distances[source][target] = new_distance

    distance_matrix = pd.DataFrame(distances)

    return distance_matrix
```

```python
def unroll_distance_matrix(df):
    """
    Unroll a distance matrix to a DataFrame in the style of the initial dataset.

    Args:
        df (pandas.DataFrame)

    Returns:
        pandas.DataFrame: Unrolled DataFrame containing columns 'id_start', 'id_end', and 'distance'.
    """
    # Write your logic here
    id_start=df.index
    id_end=df.columns
    list1=[]
    for i in id_start:
        for j in id_end:
            dict1={}
            dict1['id_start']=i
            dict1['id_end']=j
            dict1['distance']=df[i][j]
            list1.append(dict1)
            
    df_new=pd.DataFrame(list1)
    condition=df_new['id_start']!=df_new['id_end']
    df=df_new[condition]
    
    return df
```
